```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import confusion_matrix
from sklearn . metrics import mean_squared_error , r2_score
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')# ignore warnings
```

```python
data = pd.read_csv('WMT1.csv')
data_17 = data[data.Year == 2017]
week_17 = data_17[data_17.Weekday == 'Friday']
labels_17 = week_17['label'].tolist()
adj_close_17 = week_17['Adj Close'].tolist()
```

```python
acc_list = []
for w in range(5,13):
    x = np.array(range(w))
    y = adj_close_17[0:w]

    degree = 1
    weights = np. polyfit (x,y, degree)
    model_1= np. poly1d (weights)

    pre_label = []
    count = 0

    for i in range(w,len(adj_close_17)):
        temp = ''
        if model_1(i) > adj_close_17[i-1]:
            temp = 'green'
            if temp == labels_17[i]:
                count+=1
        elif model_1(i) < adj_close_17[i-1]:
            temp = 'red'
            if temp == labels_17[i]:
                count+=1
        else:
            temp = labels_17[i-1]
            if temp == labels_17[i]:
                count+=1
        pre_label.append(temp)

    acc = count/len(pre_label)
    acc_list.append(acc)
    

plt.plot(range(5,13),acc_list)
plt.xlabel('w')
plt.ylabel('acc')
plt.title('d=1')
plt.show()
```
